## Bubble Sort:

The word 'Bubble' comes from how this algorithm works, it makes the highest values 'bubble up'.

How it works:

- Go through the array, one value at a time.
- For each value, compare the value with the next value.
- If the value is higher than the next one, swap the values so that the highest value comes last.
- Go through the array as many times as there are values in the array.



#### Uses:

- Educational Purposes:
To teach the basics of sorting algorithms and the concept of swapping elements.
- Small Datasets:
Suitable for small or nearly sorted datasets.
- Stability:
It is a stable sorting algorithm, so it maintains the relative order of equal elements.

## 1a. Bubble Sort Implementation

To implement the Bubble Sort algorithm in a programming language, we need:

- An array with values to sort.

- An inner loop that goes through the array and swaps values if the first value is higher than the next value. This loop must loop through one less value each time it runs.

- An outer loop that controls how many times the inner loop must run. For an array with n values, this outer loop must run n-1 times.

The resulting code looks like this:

```python
my_array = [11,19,20,87,67,99]
n = len(my_array)

for i in range(n-1):
    for j in range(n-1):
        if my_array[j] > my_array[j+1]:
            my_array[j], my_array[j+1] = my_array[j+1], my_array[j]
            
print(my_array)
```

## 1b. Bubble Sort Implementation

To implement the Bubble Sort algorithm in a programming language, we need:

- An array with values to sort.

- An inner loop that goes through the array and swaps values if the first value is higher than the next value. This loop must loop through one less value each time it runs.

- An outer loop that controls how many times the inner loop must run. For an array with n values, this outer loop must run n-1 times.

The resulting code looks like this:

```python
my_array = [11,19,20,87,67,99]
n = len(my_array)

for i in range(n-1):
    for j in range(n-i-1):
        if my_array[j] > my_array[j+1]:
            my_array[j], my_array[j+1] = my_array[j+1], my_array[j]
            
print(my_array)
```

## 2. Bubble Sort:

```python
my_array = [7, 3, 9, 12, 10, 11]

n = len(my_array)
for i in range(n-1):
    for j in range(n-i-1): # =============================================================================================x
        if my_array[j] > my_array[j+1]:
            my_array[j], my_array[j+1] = my_array[j+1], my_array[j]
            swapped = True
        else:
            swapped = False
    if not swapped:
        break

print("Sorted array:", my_array)
```

## 3. Count Swaps in Bubble Sort:
###

```python
#array = [7, 3, 9, 12, 11, 1]
array = [20, 50, 30, 40, 10]

print('Orignal array:', array)
n = len(array)
swap = 0

for i in range(n-1):
    swaped = False   #=============================================================================================X
    for j in range(n-1):
        
        if array[j] > array[j+1]:
            array[j], array[j+1] = array[j+1], array[j]
            swaped = True
            swap += 1
            print(f'\n{array[j]} <-> {array[j+1]}')
            print(f'array: {array}...swap: {swap}')
        
    if not swaped:
        print(f'\nProgram Exited at {swap} swaps.')
        break
            
print(f'\nSorted array: {array}\nNo. of swaps: {swap}')
```

## 4. Bubble Sort for Strings
Implement Bubble Sort to sort an array of strings in alphabetical order.

**Example Input:**

#### code

my_array = ["apple", "orange", "banana", "grape", "cherry"]

#### Output:

Sorted array: ['apple', 'banana', 'cherry', 'grape', 'orange']

###

```python
array = ['apple', 'orange', 'banana', 'grape',  'cherry']

print('Orignal array:', array)
n = len(array)
swap = 0

for i in range(n-1):
    swaped = False   #----------------------X
    for j in range(n-1):
        
        if array[j] > array[j+1]:
            array[j], array[j+1] = array[j+1], array[j]
            swaped = True
            swap += 1
            print(f'\n{array[j]} <-> {array[j+1]}')
            print(f'array: {array}...swap: {swap}')
        
    if not swaped:
        print(f'\nProgram Exited at {swap} swaps.')
        break
            
print(f'\nSorted array: {array}\nNo. of swaps: {swap}')
```

## 5. Bubble Sort for 2d Array:
###

```python
array = [
    [3, 2, 1],
    [9, 5, 6],
    [8, 7, 4]
]

print('Orignal array:', array)
swap = 0

for row in array:
    
    n = len(row)

    for i in range(n-1):
        swaped = False   #----------------------X
        for j in range(n-i-1):

            if row[j] > row[j+1]:
                row[j], row[j+1] = row[j+1], row[j]
                swaped = True
                swap += 1
                print(f'\n{row[j]} <-> {row[j+1]}')
                print(f'array: {row}...swap: {swap}')

        if not swaped:
            print(f'\nProgram Exited at {swap} swaps.')
            break
            
print(f'\nSorted array: {array}\nNo. of swaps: {swap}')
```
